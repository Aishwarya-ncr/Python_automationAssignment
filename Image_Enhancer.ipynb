PIL(Pillow) is a Python Imaging Library which can add image processing capabilities to the python interpretor.
matplotlib is used for visualizing the images.
An image enhacer is developed using the support of libraries above.

```python
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image=Image.open("fortImg.jpg")
```

```python
plt.imshow(image)   #Original image
```

```python
from PIL import ImageEnhance
plt.figure(figsize=(10,10))

img1=ImageEnhance.Color(image).enhance(2.5)  #anything above value 1 enhances features and below 1 diminishes the image
img2=ImageEnhance.Contrast(image).enhance(2.5)
img3=ImageEnhance.Brightness(image).enhance(2)
img4=ImageEnhance.Sharpness(image).enhance(2)
```

```python
#plotting all the enhanced images
plt.subplot(2,2,1)
plt.imshow(img1)
plt.title("Color")
plt.subplot(2,2,2)
plt.imshow(img2)
plt.title("Contrast")
plt.subplot(2,2,3)
plt.imshow(img3)
plt.title("Brightness")
plt.subplot(2,2,4)
plt.imshow(img4)
plt.title("Sharpness")
```

```python
# diminishing features of image

img1=ImageEnhance.Color(image).enhance(0.5)  #anything above value 1 enhances features and below 1 diminishes the image
img2=ImageEnhance.Contrast(image).enhance(0.5)
img3=ImageEnhance.Brightness(image).enhance(0.5)
img4=ImageEnhance.Sharpness(image).enhance(0.5)
#plotting all the enhanced images
plt.subplot(2,2,1)
plt.imshow(img1)
plt.title("Color")
plt.subplot(2,2,2)
plt.imshow(img2)
plt.title("Contrast")
plt.subplot(2,2,3)
plt.imshow(img3)
plt.title("Brightness")
plt.subplot(2,2,4)
plt.imshow(img4)
plt.title("Sharpness")
```
